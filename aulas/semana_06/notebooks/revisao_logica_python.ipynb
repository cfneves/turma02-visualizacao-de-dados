## 🔷 Bloco 3 — Operadores

### Operadores Aritméticos (contas)

| Operador | Significado | Exemplo |
|----------|-------------|---------|
| `+` | Soma | `10 + 5 = 15` |
| `-` | Subtração | `10 - 3 = 7` |
| `*` | Multiplicação | `4 * 3 = 12` |
| `/` | Divisão | `10 / 4 = 2.5` |
| `//` | Divisão inteira | `10 // 4 = 2` |
| `%` | Resto da divisão | `10 % 3 = 1` |
| `**` | Potência | `2 ** 8 = 256` |

````python
# =============================================
# 🏭 CENÁRIO: Calculando indicadores de produção
# =============================================

pecas_produzidas = 1500
pecas_com_defeito = 37
horas_trabalhadas = 8
numero_operadores = 6
custo_por_peca = 12.50  # reais

# Peças boas
pecas_boas = pecas_produzidas - pecas_com_defeito
print(f"Peças boas: {pecas_boas}")

# Taxa de defeito (em %)
taxa_defeito = (pecas_com_defeito / pecas_produzidas) * 100
print(f"Taxa de defeito: {taxa_defeito:.2f}%")

# Produtividade por operador
produtividade = pecas_produzidas / numero_operadores
print(f"Peças por operador: {produtividade:.1f}")

# Produção por hora
pecas_por_hora = pecas_produzidas / horas_trabalhadas
print(f"Peças por hora: {pecas_por_hora:.1f}")

# Faturamento do turno
faturamento = pecas_boas * custo_por_peca
print(f"Faturamento do turno: R$ {faturamento:,.2f}")
````

````python
# Operadores de comparação — retornam True ou False
print("=== COMPARAÇÕES ===")
print(f"1500 > 1000? {1500 > 1000}")  # Maior que
print(f"37 < 50? {37 < 50}")  # Menor que
print(f"taxa_defeito <= 5? {taxa_defeito <= 5.0}")  # Menor ou igual
print(f"turno == 'Tarde'? {'Manhã' == 'Tarde'}")  # Igual
print(f"turno != 'Tarde'? {'Manhã' != 'Tarde'}")  # Diferente
````

Condicionais permitem que o programa **tome decisões** com base em uma condição.

```python
if condicao_verdadeira:
```

> 🔑 **A indentação (espaços no início da linha) é OBRIGATÓRIA em Python!**

````python
# =============================================
# 🏭 CENÁRIO: Sistema de alertas de qualidade
# =============================================

taxa_defeito = 7.2  # tente mudar esse valor!

print(f"Taxa de defeito atual: {taxa_defeito}%")
print()

if taxa_defeito <= 2.0:
    status = "✅ EXCELENTE"
    acao = "Manter padrão atual"
elif taxa_defeito <= 5.0:
    status = "⚠️  ATENÇÃO"
    acao = "Monitorar de perto e revisar processo"
elif taxa_defeito <= 10.0:
    status = "🔴 CRÍTICO"
    acao = "Acionar supervisor imediatamente"
else:
    status = "🚨 PARAR LINHA"
    acao = "Interromper produção e chamar engenharia"

print(f"Status: {status}")
print(f"Ação recomendada: {acao}")
````

````python
# Usando operadores lógicos: and, or, not

temperatura = 195.0
pressao = 3.2
valvula_aberta = True

# and = as DUAS condições precisam ser verdadeiras
if temperatura > 180 and pressao > 3.0:
    print("⚙️  Processo dentro do padrão")

# or = PELO MENOS UMA condição precisa ser verdadeira
if temperatura > 220 or pressao > 5.0:
    print("🚨 Risco de superaquecimento!")
else:
    print("✅ Temperatura e pressão normais")

# not = inverte o valor lógico
if not valvula_aberta:
    print("❌ Abra a válvula antes de iniciar!")
else:
    print("✅ Válvula aberta — pode iniciar")
````

## 🔷 Bloco 6 — Estruturas de Repetição

### 6.1 — Loop `for`
Usamos `for` quando sabemos **quantas vezes** queremos repetir algo.

````python
# =============================================
# 🏭 CENÁRIO: Relatório diário de produção
# =============================================

# Lista de produção por dia da semana
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]

print("=== PRODUÇÃO SEMANAL ===")

total = 0
for i in range(len(dias)):
    print(f"{dias[i]}: {producao_semanal[i]} peças")
    total += producao_semanal[i]  # acumula o total

media = total / len(dias)
print(f"\nTotal da semana: {total} peças")
print(f"Média diária: {media:.1f} peças")
````

````python
# Usando range() — gera sequências de números
print("Contagem de 1 a 5:")
for numero in range(1, 6):
    print(numero)

print("\nTabuada do 3:")
for i in range(1, 11):
    resultado = 3 * i
    print(f"3 x {i:2} = {resultado:3}")
````

````python
# for com if dentro — combinando tudo!
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]
meta_diaria = 1500

print("=== ANÁLISE DA META ===")
dias_acima = 0
dias_abaixo = 0

for i in range(len(dias)):
    producao = producao_semanal[i]
    if producao >= meta_diaria:
        print(f"✅ {dias[i]}: {producao} — META BATIDA")
        dias_acima += 1
    else:
        diferenca = meta_diaria - producao
        print(f"❌ {dias[i]}: {producao} — faltaram {diferenca} peças")
        dias_abaixo += 1

print(f"\nDias acima da meta: {dias_acima}")
print(f"Dias abaixo da meta: {dias_abaixo}")
````

### 6.2 — Loop `while`
Usamos `while` quando queremos repetir **enquanto uma condição for verdadeira**.

> ⚠️ **Cuidado com loops infinitos!** A condição precisa se tornar falsa em algum momento.

````python
# =============================================
# 🏭 CENÁRIO: Simulando uma esteira de produção
# =============================================

estoque_materias_primas = 100  # kg de matéria-prima
consumo_por_lote = 15  # kg consumidos por lote
lotes_produzidos = 0

print("=== SIMULAÇÃO DE PRODUÇÃO ===")
print(f"Estoque inicial: {estoque_materias_primas} kg\n")

while estoque_materias_primas >= consumo_por_lote:
    estoque_materias_primas -= consumo_por_lote
    lotes_produzidos += 1
    print(
        f"Lote {lotes_produzidos} produzido | Estoque restante: {estoque_materias_primas} kg"
    )

print(f"\n⚠️  Matéria-prima insuficiente para novo lote")
print(f"Total de lotes produzidos: {lotes_produzidos}")
print(f"Sobra de material: {estoque_materias_primas} kg")
````

## 🔷 Bloco 8 — Funções

Funções são **blocos de código reutilizáveis**. Em vez de repetir o mesmo código, você cria uma função e chama ela quando precisar.

```python
def nome_da_funcao(parametro1, parametro2):
    # código aqui
    return resultado
```

````python
# =============================================
# 🏭 CENÁRIO: Funções para indicadores de qualidade
# =============================================


def calcular_taxa_defeito(total_produzido, total_defeituoso):
    """Calcula a taxa de defeito em porcentagem."""
    if total_produzido == 0:
        return 0
    taxa = (total_defeituoso / total_produzido) * 100
    return taxa


def classificar_qualidade(taxa_defeito):
    """Classifica a qualidade com base na taxa de defeito."""
    if taxa_defeito <= 1.0:
        return "EXCELENTE"
    elif taxa_defeito <= 3.0:
        return "BOM"
    elif taxa_defeito <= 5.0:
        return "ATENÇÃO"
    else:
        return "CRÍTICO"


def gerar_relatorio_turno(turno, operador, pecas, defeitos):
    """Gera um relatório completo do turno."""
    taxa = calcular_taxa_defeito(pecas, defeitos)
    classificacao = classificar_qualidade(taxa)
    pecas_boas = pecas - defeitos

    print(f"\n{'=' * 40}")
    print(f"  RELATÓRIO — TURNO {turno.upper()}")
    print(f"{'=' * 40}")
    print(f"  Operador      : {operador}")
    print(f"  Peças prod.   : {pecas}")
    print(f"  Peças boas    : {pecas_boas}")
    print(f"  Defeituosas   : {defeitos}")
    print(f"  Taxa defeito  : {taxa:.2f}%")
    print(f"  Classificação : {classificacao}")
    print(f"{'=' * 40}")


# Chamando as funções
gerar_relatorio_turno("Manhã", "Ana Silva", 520, 8)
gerar_relatorio_turno("Tarde", "Bruno Costa", 495, 28)
gerar_relatorio_turno("Noite", "Carla Souza", 485, 4)
````

### ✏️ Atividade Acompanhada 4 — Crie sua função
Crie uma função que calcule o OEE (Overall Equipment Effectiveness) — um dos indicadores mais importantes na indústria:

````python
# ✏️ ATIVIDADE ACOMPANHADA 4
# OEE = Disponibilidade × Performance × Qualidade × 100
# Cada fator vai de 0 a 1 (ex: 0.85 = 85%)


def calcular_oee(disponibilidade, performance, qualidade):
    """
    Calcula o OEE (Overall Equipment Effectiveness)

    Parâmetros:
        disponibilidade: tempo disponível / tempo planejado (0 a 1)
        performance: velocidade real / velocidade ideal (0 a 1)
        qualidade: peças boas / total produzido (0 a 1)
    """
    # TODO: calcule o OEE
    oee = (
        disponibilidade * performance * qualidade
    ) * 100  # substitua pelo cálculo correto
    return oee


# Testando a função
oee_resultado = calcular_oee(
    disponibilidade=0.90,  # máquina disponível 90% do tempo
    performance=0.95,  # operando a 95% da velocidade ideal
    qualidade=0.98,  # 98% de peças boas
)

print(f"OEE calculado: {oee_resultado:.1f}%")

# OEE de classe mundial é ≥ 85%
if oee_resultado >= 85:
    print("🏆 Equipamento de CLASSE MUNDIAL!")
else:
    print("📈 Há oportunidade de melhoria")
````

````python
# ============================================================
# 📝 ATIVIDADE INDIVIDUAL 2 — Verificador de Lote
# ============================================================
#
# ENUNCIADO:
# Um lote de peças passa pelo controle de qualidade.
# Você tem a lista de medições de diâmetro (em mm) abaixo.
# A especificação permite entre 49.8mm e 50.2mm.
#
# Crie um programa que:
# 1. Percorra a lista de medições
# 2. Conte quantas peças estão DENTRO e FORA da especificação
# 3. Calcule o percentual de aprovação
# 4. Classifique o lote:
#    - Aprovado: aprovação >= 98%
#    - Aprovado c/ ressalvas: entre 95% e 97.9%
#    - Reprovado: < 95%
# ============================================================

medicoes = [
    50.1,
    49.9,
    50.0,
    50.3,
    49.7,
    50.2,
    50.1,
    49.8,
    50.0,
    50.1,
    49.6,
    50.2,
    50.0,
    49.9,
    50.1,
    50.4,
    49.8,
    50.0,
    50.1,
    49.9,
]

especificacao_min = 49.8
especificacao_max = 50.2

# SEU CÓDIGO AQUI:

print(f"\n{'=' * 40}\n  VERIFICADOR DE LOTE\n{'=' * 40}")

dentro_especificacao = 0
fora_especificacao = 0

for medida in medicoes:
    if especificacao_min <= medida <= especificacao_max:
        dentro_especificacao += 1
    else:
        fora_especificacao += 1
total_peca = len(medicoes)
percentual_aprovacao = (dentro_especificacao / total_peca) * 100

print("==== RELATÓRIO DE LOTE ====")
print(f"Peças dentro das especificações: {dentro_especificacao}")
print(f"Fora daa especificações: {fora_especificacao}")
print(f"Percentual de Aprovação: {percentual_aprovacao:.2f}%")

if percentual_aprovacao >= 98:
    print("Lote Aprovado")
elif percentual_aprovacao >= 95:
    print("Lote Aprovado com ressalvas")
else:
    print("Lote Reprovado")
````

````python
# ============================================================
# SISTEMA DE REGISTRO DE PRODUÇÃO
# ============================================================

# Criamos uma lista vazia que vai guardar todos os registros.
# Cada registro será um dicionário (como uma ficha de dados).
registros = []


# ============================================================
# FUNÇÃO 1 — Registrar produção de um turno
# ============================================================
# "def" cria uma função: um bloco de código com nome próprio
# que podemos chamar várias vezes ao longo do programa.
def registrar_producao():

    print("\n--- REGISTRAR PRODUÇÃO ---")
    # "\n" pula uma linha antes do texto, deixando mais legível.

    # input() exibe uma mensagem e espera o usuário digitar algo.
    # O que for digitado fica salvo na variável à esquerda do "=".
    nome = input("Nome do operador: ")

    # Aqui pedimos o turno e já deixamos a dica dos valores aceitos.
    turno = input("Turno (Manhã/Tarde/Noite): ")

    # int() converte o texto digitado para número inteiro,
    # porque input() sempre retorna texto (string).
    quantidade = int(input("Quantidade produzida: "))

    defeitos = int(input("Quantidade com defeito: "))

    # Calculamos a taxa de defeito automaticamente.
    # Para evitar divisão por zero (erro quando quantidade == 0),
    # usamos um "if" inline (operador ternário):
    # se quantidade > 0, calcula; senão retorna 0.0
    taxa_defeito = (defeitos / quantidade * 100) if quantidade > 0 else 0.0

    # Criamos um dicionário com todos os dados do registro.
    # Dicionários têm pares "chave": valor — como campos de um formulário.
    registro = {
        "operador": nome,
        "turno": turno,
        "quantidade": quantidade,
        "defeitos": defeitos,
        "taxa_defeito": taxa_defeito,  # valor já calculado acima
    }

    # .append() adiciona o dicionário recém-criado ao final da lista.
    registros.append(registro)

    # f-string (f"...") permite colocar variáveis dentro do texto
    # usando chaves {}. O :.2f formata o número com 2 casas decimais.
    print(f"\nRegistro salvo! Taxa de defeito: {taxa_defeito:.2f}%")


# ============================================================
# FUNÇÃO 2 — Ver todos os registros
# ============================================================
def ver_registros():

    print("\n--- TODOS OS REGISTROS ---")

    # Verificamos se a lista está vazia antes de tentar exibir.
    # "not registros" é True quando a lista não tem nenhum item.
    if not registros:
        print("Nenhum registro encontrado.")
        return  # "return" encerra a função imediatamente aqui.

    # enumerate() percorre a lista e entrega dois valores a cada volta:
    # o índice (posição) e o item em si.
    # Começamos em 1 (start=1) para a numeração ficar mais natural.
    for i, r in enumerate(registros, start=1):
        # Separador visual entre registros
        print(f"\nRegistro #{i}")
        print(f"  Operador : {r['operador']}")
        print(f"  Turno    : {r['turno']}")
        print(f"  Produção : {r['quantidade']} unidades")
        print(f"  Defeitos : {r['defeitos']} unidades")
        # {r['taxa_defeito']:.2f} acessa a chave 'taxa_defeito' do dicionário
        # e formata com 2 casas decimais.
        print(f"  Tx. def. : {r['taxa_defeito']:.2f}%")
        print("  " + "-" * 30)  # linha divisória decorativa


# ============================================================
# FUNÇÃO 3 — Ver estatísticas
# ============================================================
def ver_estatisticas():

    print("\n--- ESTATÍSTICAS ---")

    # Mesma verificação: não faz sentido calcular com lista vazia.
    if not registros:
        print("Nenhum registro para calcular estatísticas.")
        return

    # Usamos "list comprehension" para extrair só as quantidades.
    # É equivalente a um for que monta uma lista nova:
    # quantidades = []
    # for r in registros:
    #     quantidades.append(r["quantidade"])
    quantidades = [r["quantidade"] for r in registros]

    # Mesmo processo para as taxas de defeito.
    taxas = [r["taxa_defeito"] for r in registros]

    # sum() soma todos os valores da lista.
    total = sum(quantidades)

    # max() e min() retornam o maior e menor valor, respectivamente.
    maior = max(quantidades)
    menor = min(quantidades)

    # len() retorna a quantidade de itens na lista.
    # Dividimos a soma pelo total de registros para obter a média.
    media = total / len(quantidades)

    media_defeito = sum(taxas) / len(taxas)

    print(f"Total produzido    : {total} unidades")
    print(f"Maior produção     : {maior} unidades")
    print(f"Menor produção     : {menor} unidades")
    print(f"Média de produção  : {media:.2f} unidades")
    print(f"Média de defeitos  : {media_defeito:.2f}%")


# ============================================================
# PROGRAMA PRINCIPAL — Loop do menu
# ============================================================

# Exibimos o título uma única vez antes do loop começar.
print("=" * 40)
print("  SISTEMA DE REGISTRO DE PRODUÇÃO")
print("=" * 40)

# "while True" cria um loop infinito — o programa fica rodando
# até encontrarmos um "break" dentro dele.
while True:
    # Exibimos o menu a cada volta do loop.
    print("\n--- MENU ---")
    print("[1] Registrar produção de um turno")
    print("[2] Ver todos os registros")
    print("[3] Ver estatísticas")
    print("[4] Sair")

    # Lemos a opção digitada pelo usuário.
    opcao = input("\nEscolha uma opção: ")

    # Comparamos a string digitada com cada opção possível.
    if opcao == "1":
        registrar_producao()  # chama a função de registro

    elif opcao == "2":
        ver_registros()  # chama a função de listagem

    elif opcao == "3":
        ver_estatisticas()  # chama a função de estatísticas

    elif opcao == "4":
        # Mensagem de despedida e "break" encerra o while True.
        print("\nEncerrando o sistema. Até logo!")
        break

    else:
        # Se o usuário digitar qualquer outra coisa, avisamos
        # e o loop continua normalmente (volta pro menu).
        print("Opção inválida! Digite 1, 2, 3 ou 4.")
````
