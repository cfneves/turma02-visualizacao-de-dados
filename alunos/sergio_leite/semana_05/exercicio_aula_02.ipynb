```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
URL = (
    "https://raw.githubusercontent.com/"
    "cfneves/turma-visualizacao-de-dados/"
    "master/aulas/semana_04/bases/base_rh.xlsx"
)

df = pd.read_excel(URL) 

# conversão datetime
df["Data_Admissao"] = pd.to_datetime(
    df["Data_Admissao"],
    format="%d/%m/%Y",
    errors="coerce"
)

```

```python
df["Ano_Admissao"] = df["Data_Admissao"].dt.year # extrai o ano de admissão
print(df[["Data_Admissao", "Ano_Admissao"]].head())
```

```python
# 2. contagem de funcionarios por departamento e cargo
contagem_depto_cargo = df.groupby(["Departamento", "Cargo"]).size().reset_index(name="Total_Funcionarios")

# Resultado
display(contagem_depto_cargo)
```

```python
# Contagem de admissões por ano 
admissoes_por_ano = df.groupby("Ano_Admissao").size().reset_index(name="Contratacoes")

# filtro com query para filtrar o intervalo de 2020 a 2024
admissoes_filtradas = admissoes_por_ano.query("2020 <= Ano_Admissao <= 2024")

display(admissoes_filtradas)
```

```python
# 4a. Calcular o Headcount atual real por departamento
headcount_atual = df.groupby("Departamento").size().reset_index(name="Total_Atual")

# 4b. Criar a tabela de metas fictícias baseada nos departamentos existentes
departamentos_unicos = headcount_atual["Departamento"].tolist()
# Criando metas exemplo (ex: 50 para cada depto ou valores variados)
metas_dados = {
    "Departamento": departamentos_unicos,
    "Meta_Headcount": [100, 120, 80, 150, 90, 110, 95][:len(departamentos_unicos)] # Ajusta ao tamanho da lista
}
df_metas = pd.DataFrame(metas_dados)

# 4c. Fazer o merge left
df_analise_metas = pd.merge(df_metas, headcount_atual, on="Departamento", how="left")

# 4d. Identificar se a meta foi atingida
df_analise_metas["Meta_Atingida"] = df_analise_metas["Total_Atual"] >= df_analise_metas["Meta_Headcount"]

display(df_analise_metas)
```
