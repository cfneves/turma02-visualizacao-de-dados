```python
# Manipulação de datas - Módulo datetime
# Quando importamos dados de arquivos CSV, excel ou Json, as datas sempre entram
# no python como se fossem texto comum (string) e não uma data. 
# A coluna data de admissão foi criada como uma string
# strptime - string parse time - transforma texto para date
# strftime - string format time - transforma data para texto
# Códigos de formato da tela:
    # %d: Dia com dois dígitos (ex: 08)
    # %m: Mês com dois dígitos (ex: 08)
    # %Y: (Y MAIUSCULO) Ano com 4 dígitos (ex: 2024) -> Nota: Y maiúsculo é para 4 dígitos, y minúsculo seria apenas para 2 dígitos (24).
    # %B: Nome do mês por extenso (ex: Agosto)

# importação da biblioteca
from datetime import datetime, timedelta

# o primeiro valor da coluna data de admissão será o exemplo
data_texto = '13/08/2024' 
# datetime.strptime(texto, formato)
# %d/%m/%y

# de TEXTO para DATA - strptime
#criamos a var para receber o datetime com o modelo do texto e o formato da data
data_obj = datetime.strptime(data_texto, '%d/%m/%Y')
print(f'String original: {data_texto}')
print(f'Objeto datetime: {data_obj}')
print(f'O tipo da data é: {type(data_obj)}')
print() # pular 1 linha

# de Data para TEXTO - strftime
# vamos formatar o texto na forma que quisermos
formatado = data_obj.strftime('%d de %B de %Y')
print(f'A data formatada ficou assim: {formatado}')

# extraindo parte da data
print(f"Dia {data_obj.day}")
print(f"Mês {data_obj.month}")
print(f"Ano {data_obj.year}")

```

```python
# timesdelta - É o resultado da subtração de duas datas, a diferença de tempo.
# se subtrair as datas, recebemos o timedelta como resultado

# datetime.today
hoje = datetime.today()
print(hoje)
print()

diferenca = hoje - data_obj # calculo do timedelta usando o valor atribuido a var data_obj
print(f'A diferenca entre a data de hoje {hoje} e a data \n  {data_obj} é igual a {hoje}')
print()
print(f'Data de admissão: {data_texto}')
print(f'Hoje é dia: {hoje.strftime("%d/%m/%Y")}') # usando o strftime ele converta a data para o formato desejado
print(f'A diferença entre as datas é de: {diferenca.days} dias') # ao colocar o .days ele vai mostrar apenas os dias
print(f'A diferença em anos é de {diferenca.days / 365.25:.1f} anos') # para criar anos, temos que usar os dias como base e calcular com o .25 por conta dos anos bissestos
```

```python
# Convertendo uma coluna inteira com o pd.to.datetime
# Usado para converser muitas linhas ao mesmo tempo.
# O exemplo anterior converteu apenas 1 linha

print(f'Tipo da data ANTES: {df['Data_Admissao'].dtype}')
print(f'Valor da data ANTES: {df['Data_Admissao'].iloc[0]}') # iloc[0] acessa a 1ª linha do indice
print()

# Agora vamos converter toda a coluna e usar uma cópia  para nao alterar o DF original
df_limpo = df.copy() # Criamos uma copia independente do DF na var df_limpo para trabalhar sem mexer no original
df_limpo['Data_Admissao'] = pd.to_datetime( 
    df_limpo['Data_Admissao'],
    format='%d/%m/%Y', # define o formato que queremos para a string de entrada
    errors='coerce' # define que datas invalidas viram Nat (Not a Time) e nao erro
)

print(f'O valor NOVO é: {df_limpo['Data_Admissao'].iloc[0]}') # imprime o df na data de admissão e no indice 0
print(f'O tipo NOVO é: {df_limpo["Data_Admissao"].dtype}')
print(f'Os Nats gerados foram: {df_limpo["Data_Admissao"].isna().sum()}')  # isna() vai listar todos os NaTs e o sum vai somar e dar um resoltado final
```

```python
# Extraindo informações com o acessor .dt.
# Depois de converter para o datetime p pandas libera o acessor .dt para acessar as datas
# .dt.year - retorna o ano (2026)
# .dt.month - retorna o mês (10)
# .dt.quarter - retorna o trimestre (3)
# .dt.day_name() - retorna o dia da semana ("monday")

# criando colunas derivadas da data 
df_limpo['Ano_Admissao'] = df_limpo['Data_Admissao'].dt.year
df_limpo['Mes_Admissao'] = df_limpo['Data_Admissao'].dt.month
df_limpo['Trimestre'] = df_limpo['Data_Admissao'].dt.quarter

# print(df_limpo['Ano_Admissao']) # testando a coluna criada

# calculando o tempo de casa em anos
# print(pd.Timestamp.today()) # comando para imprimir a data e hora de hoje no pandas

# calculando o tempo de casa em anos
hoje_ts = pd.Timestamp.today() # atribui a data de hoje na var


# criamos a tabela tempo_casa_anos e atribuimos a ela o calculo da data de:
#  hoje (hoje_ts) - o acesso a data de admissão em dias. Dividimos por 365.25 para considerar o ano bisexto
df_limpo['Tempo_Casa_Anos'] = (
    (hoje_ts - df_limpo['Data_Admissao']).dt.days / 365.25).round(2)
print(df_limpo['Tempo_Casa_Anos'].iloc[2])

# montando uma lista com o resultado
cols = ['Nome', 'Data_Admissao', 'Ano_Admissao', 'Mes_Admissao', 'Trimestre', "Tempo_Casa_Anos"]
df_limpo[cols].head(5)


```

```python
# Uando função
# def nome_funcao(parametro):
#    return resultado

# construindo uma função para classificar o tempo de casa em uma categoria de texto
# é interessante criar categoriar pois é mais fácil para utilizar com gráficos e filtros no power BI do que números decimais

def classificar_tempo(anos):
    """
    Recebe: Anos (número) e o tempo de casa do funcionário
    Retorna: Uma categoria em texto
    """
    # pd.isna() - verifica se o valor é nulo (NaN ou NaT)
    if pd.isna(anos):
        return 'Sem data'
    elif anos < 1:
        return 'Menos de 1 ano'
    elif anos < 3:
         return '1 a 3 anos'
    elif anos < 5:
        return '3 a 5 anos'
    else:
        return 'Mais de 5 anos'
    
# teste da função com valores individuais 
print(classificar_tempo(7))
```

```python
# Aplicando a função em cada linha da coluna
# Fazendo loop em cada celular de forma eficiente

df_limpo['Faixa_Tempo_Casa'] = df_limpo['Tempo_Casa_Anos'].apply(classificar_tempo)

print('Distribuição por faixa de tempo na casa')
print(df_limpo['Faixa_Tempo_Casa'].value_counts())
```
