```python
#dt.isnull() cria um dataframe booleano, com True para cada valor nulo, e False para cada valor não nulo. dt.isnull().sum() soma os valores booleanos, contando quantos valores nulos tem em cada coluna.
#.sum() soma os valores booleanos, contando quantos valores nulos tem em cada coluna.
#resultado: quantidade de valores nulos em cada coluna, para confirmar que os dados foram lidos corretamente, e para entender a estrutura do dataframe.
print('Quantidade de valores nulos em cada coluna:\n', df.isnull().sum()) #quantidade de valores nulos em cada coluna, para confirmar que os dados foram lidos corretamente, e para entender a estrutura do dataframe.

#percentual de nulo - mais util para entender a proporção de nulos em cada coluna, especialmente se o número total de linhas for grande. df.isnull().mean() calcula a média dos valores booleanos, dando a proporção de valores nulos em cada coluna.
# len(df) #número total de linhas do dataframe, usado para calcular o percentual de nulos.
#resultado: percentual de valores nulos em cada coluna, para confirmar que os dados foram lidos corretamente, e para entender a estrutura do dataframe.
print('Percentual de valores nulos em cada coluna:\n', df.isnull().mean()) #percentual de valores nulos em cada coluna, para confirmar que os dados foram lidos corretamente, e para entender a estrutura do dataframe.
```

```python
#filtro composto: df[(df['Status'] == 'Ativo') & (df['Departamento'] == 'Vendas')] #filtra as linhas onde a coluna "Status" é igual a "Ativo" e a coluna "Departamento" é igual a "Vendas", mostrando apenas os funcionários ativos do departamento de vendas. Use & para "e", | para "ou", e ~ para "não". Lembre-se de colocar cada condição entre parênteses.

#use & para "e", | para "ou", e ~ para "não". Lembre-se de colocar cada condição entre parênteses. Exemplo: df[(df['Status'] == 'Ativo') & (df['Departamento'] == 'Vendas')] #filtra as linhas onde a coluna "Status" é igual a "Ativo" e a coluna "Departamento" é igual a "Vendas", mostrando apenas os funcionários ativos do departamento de vendas.

gerentes_ti = df[(df['Cargo'] == 'Gerente') & (df['Departamento'] == 'TI')] #filtra as linhas onde a coluna "Cargo" é igual a "Gerente" e a coluna "Departamento" é igual a "TI", mostrando apenas os gerentes do departamento de TI.
print(f'Número de gerentes de TI: {len(gerentes_ti)}') #imprime o número de gerentes de TI, contando as linhas do dataframe filtrado. Substitua 'Cargo' e 'Departamento' pelos nomes das colunas que você quer filtrar, e 'Gerente' e 'TI' pelos valores que você quer filtrar. 
gerentes_ti[['Nome', 'Salario','Status']] #mostra as colunas "Nome" e "Salario" do dataframe filtrado, para ver os nomes e salários dos gerentes de TI. Substitua 'Nome' e 'Salario' pelos nomes das colunas que você quer mostrar.
```

```python
#Estátiticas rapidas de uma coluna numérica: df['coluna'].mean(), df['coluna'].median(), df['coluna'].min(), df['coluna'].max(), df['coluna'].std() #mostra a média, mediana, mínimo, máximo, e desvio padrão da coluna, para entender a distribuição dos dados, e para confirmar que os dados foram lidos corretamente. Substitua 'coluna' pelo nome da coluna que você quer analisar.
#mean() mostra a média da coluna, para entender a distribuição dos dados, e para confirmar que os dados foram lidos corretamente. Substitua 'coluna' pelo nome da coluna que você quer analisar.
#max() mostra o valor máximo da coluna, para entender a distribuição dos dados, e para confirmar que os dados foram lidos corretamente. Substitua 'coluna' pelo nome da coluna que você quer analisar.
#min() mostra o valor mínimo da coluna, para entender a distribuição dos dados, e para confirmar que os dados foram lidos corretamente. Substitua 'coluna' pelo nome da coluna que você quer analisar.
#mediana() mostra a mediana da coluna, para entender a distribuição dos dados, e para confirmar que os dados foram lidos corretamente. Substitua 'coluna' pelo nome da coluna que você quer analisar.
                             
print(f'Média do salário: {df["Salario"].mean():.2f}')
print(f'Máximo do salário: {df["Salario"].max():.2f}')
print(f'Mínimo do salário: {df["Salario"].min():.2f}')
print(f'Mediana do salário: {df["Salario"].median():.2f}')
```
