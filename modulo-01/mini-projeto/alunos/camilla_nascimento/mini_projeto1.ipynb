2. TRANSFORMAÇÃO DE DADOS

O objetico é padronizar os ditpos de dados para permitir a análise correta e as futuras operações estatísticas.

```python
print("\n" + "=" * 50)
print("SPRINT 2 - TRANSFORMAÇÃO DOS DADOS")
print("=" * 50)


df_tratado = df.copy()  # Cria uma cópia do DataFrame original para realizar as transformações sem alterar os dados originais.

df_tratado["DATA"] = pd.to_datetime(
    df_tratado["DATA"],
    dayfirst=True,
    errors="coerce"
)  # Converte a coluna "DATA" para o formato datetime, considerando o formato dia/mês/ano. Se houver erros na conversão, os valores problemáticos serão convertidos para NaT (Not a Time).

print("\nColuna DATA convertida para datetime com sucesso.")

datas_invalidas = df_tratado["DATA"].isna().sum()
print(f"\nQuantidade de datas inválidas encontradas: {datas_invalidas}")

colunas_texto = [
    "CL_GENERO",
    "CL_SEG",
    "PR_CAT",
    "PR_NOME"
]

for coluna in colunas_texto:
    df_tratado[coluna] = (
        df_tratado[coluna]
        .astype(str)
        .str.strip()
        .str.upper()
    )

print("Padronização dos campos de texto concluída.")


print("\nTipos de dados após as transformações:")

print(df_tratado.dtypes)

print(df["DATA"].dtype)
print(df_tratado["DATA"].dtype)
```

Dessas atividades foram executadas:
1. criação de uma cópia da base original para preservar os dados;
2. conversão da coluna DATA, que estava armazenada como texto, com o código datetime;
3. verificação de possíveis datas inválidas;
4. padronização dos campos textuais através da remoção de espaços e a conversão para letras maiúsculas para manter um padrão;
5. conferência dos tipos de dados após as transformações.

A coluna DATA foi convertida com sucesso e não foram encontradas datas inválidas, bem como os campos textuais foram padronizados.


3. DESENVOLVIMENTO DAS FUNÇÕES DE LIMPEZA E TRATAMENTO DE DADOS

NO objetivo é melhorar a qualidade da base de dados removendo informações desnecessárias e registros repetidos ou inconsistentes. 

```python
# LIMPEZA E TRATAMENTO - PADRONIZANDO OS DADOS PARA ANÁLISE

print("\n" + "=" * 50)
print("SPRINT 3 - LIMPEZA DE NULOS E DUPLICATAS")
print("=" * 50)

# Verificando as colunas vazias
colunas_vazias = df_tratado.columns[df_tratado.isnull().all()]
print("\nColunas completamente vazias encontradas:")
for coluna in colunas_vazias:
    print(f"- {coluna}")
df_tratado = df_tratado.drop(columns=colunas_vazias)
print("\nColunas vazias removidas com sucesso.")

#Verificando os valores nulos 
print("\nQuantidade de valores nulos por coluna:")
nulos = df_tratado.isnull().sum()
print(nulos)

#Tratando os valores nulos
total_nulos = nulos.sum()

if total_nulos == 0:
    print("\nNenhum valor nulo encontrado nas colunas úteis da base.")
else:
    print(f"\nForam encontrados {total_nulos} valores nulos.")


#Verificando os registros duplicados
duplicados = df_tratado.duplicated().sum()   #NECESSÁRIO REMOVER OS DUPLICADOS, PODEM SER DUAS COMPRAS.. VERIFICAR
print(f"\nQuantidade de registros duplicados: {duplicados}")


#Remoção dos duplicados
if duplicados > 0:
    
    linhas_antes = len(df_tratado)

    df_tratado = df_tratado.drop_duplicates()

    linhas_depois = len(df_tratado)

    removidos = linhas_antes - linhas_depois

    print(f"Foram removidos {removidos} registros duplicados.")

else:
    
    print("Nenhum registro duplicado foi encontrado.")                                

```

```python
print("Linhas originais:", len(df))
print("Linhas após limpeza:", len(df_tratado))

print(df_tratado[df_tratado.duplicated()].shape)
```

Atividades executadas:
1. identificação e remoção das colunas completamente vazias;
2. verificação da existência de valores nulos nas colunas;
3. análise da quantidade de registros duplicados;
4. remoção dos registros considerados duplicados.

Obs.: Nesse tópico cabe uma atenção nos dados duplicados, tendo em vista que no caso da base analisada pode ser um dado que realmente pode ser válido, a remoção foi feita, mas com a necessidade de verificação posterior se relamente esses dados são duplicados ou são válidos. Como os registros apresentavam igualdade em todas as colunas da base, optou-se pela remoção dessas duplicidades para evitar distorções nas análises estatísticas e nos agrupamentos posteriores.

Foram removidas 4 colunas vazias e não foram encontrados valores nulos nas colunas úteis, foram removidos 96.553 registros duplicados e a base passou a conter 733.447 registros válidos para a referida análise.

4. SPRINT DE ESTASTÍSTICA DESCRITIVA

O objetivo é aplicar técnicas de estatística descritiva para compreender o comportamento da variável referente ao número de filhos dos clientes.

O objetivo é identificar padrões de comportamento por meio do agrupamentos dos gêneros.
