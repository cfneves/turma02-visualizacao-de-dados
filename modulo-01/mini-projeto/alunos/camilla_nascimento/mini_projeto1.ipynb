1. SPRINT 1 - IMPORTAÇÃO DE DADOS
Nesse tópico é estabelecida a base de dados a ser usada (kaggle), conforme determina o projeto. A base de dados está disponível no módulo-01 - base_do_projeto - Base Varejo.csv. Dessa forma importamos o csv e por fim a importação do pandas para criar as tabelas e os relatórios para as análises. Inicialmente importamos os arquivos e carregamos a base de dados e verificamos se a leitura funcionou.

```python
# Importando as bibliotecas necessárias para o projeto 

import csv
from datetime import datetime
import pandas as pd

caminho_arquivo = r"C:\Users\Milla\Documents\GitHub\turma-visualizacao-de-dados\modulo-01\base_do_projeto\Base Varejo.csv"

df = pd.read_csv(caminho_arquivo, sep=";", encoding="cp1252", decimal=",")

print("Arquivo csv carregado com sucesso!")

print("\nPrimeiras 5 linhas:")
print(df.head())

# Informações gerais sobre o DataFrame

print(f"Quantidade de registros: {df.shape[0]:,}")                  # Mostra a quantidade de registros (linhas) presentes no DataFrame, formatada com vírgula para melhor leitura.
print(f"Quantidade de colunas: {df.shape[1]:,}")                    # Mostra a quantidade de colunas presentes no DataFrame, formatada com vírgula para melhor leitura. 

for coluna in df.columns:
    print(f"Coluna: {coluna} - Tipo de dado: {df[coluna].dtype}")   # Itera sobre cada coluna do DataFrame e imprime o nome da coluna junto com o tipo de dado que ela contém.

print(df.dtypes)

nulos = df.isnull().sum()                                            # Calcula a quantidade de valores nulos em cada coluna do DataFrame.
print("\nQuantidade de valores nulos por coluna:")
print(nulos)                                                         # Imprime a quantidade de valores nulos por coluna. 

duplicados = df.duplicated().sum()                                   # Calcula a quantidade de linhas duplicadas no DataFrame.
print(f"\nQuantidade de linhas duplicadas: {duplicados}")            # Imprime a quantidade de linhas duplicadas encontradas no DataFrame. 

colunas_vazias = df.columns[df.isnull().all()]                       # Identifica as colunas que estão completamente vazias (todas as linhas são nulas).
print(f"\nColunas completamente vazias: {colunas_vazias.tolist()}")  # Imprime a lista de colunas que estão completamente vazias, convertendo o resultado para uma lista legível.

if len(colunas_vazias) > 0:
    for coluna in colunas_vazias:
        print(f"Coluna '{coluna}' está completamente vazia e pode ser removida.")  # Informa quais colunas estão completamente vazias e sugere que podem ser removidas.
else:
    print("\nNenhuma coluna completamente vazia encontrada.")        # Informa que não foram encontradas colunas completamente vazias.

```

2. TRANSFORMAÇÃO DE DADOS



```python
print("\n" + "=" * 50)
print("SPRINT 2 - TRANSFORMAÇÃO DOS DADOS")
print("=" * 50)


df_tratado = df.copy()  # Cria uma cópia do DataFrame original para realizar as transformações sem alterar os dados originais.

df_tratado["DATA"] = pd.to_datetime(
    df_tratado["DATA"],
    dayfirst=True,
    errors="coerce"
)  # Converte a coluna "DATA" para o formato datetime, considerando o formato dia/mês/ano. Se houver erros na conversão, os valores problemáticos serão convertidos para NaT (Not a Time).

print("\nColuna DATA convertida para datetime com sucesso.")

datas_invalidas = df_tratado["DATA"].isna().sum()
print(f"\nQuantidade de datas inválidas encontradas: {datas_invalidas}")

colunas_texto = [
    "CL_GENERO",
    "CL_SEG",
    "PR_CAT",
    "PR_NOME"
]

for coluna in colunas_texto:
    df_tratado[coluna] = (
        df_tratado[coluna]
        .astype(str)
        .str.strip()
        .str.upper()
    )

print("Padronização dos campos de texto concluída.")


print("\nTipos de dados após as transformações:")

print(df_tratado.dtypes)

print(df["DATA"].dtype)
print(df_tratado["DATA"].dtype)
```

2. SPRINT 3 - DESENVOLVIMENTO DAS FUNÇÕES DE LIMPEZA E TRATAMENTO DE DADOS

Nesse momento vai ser feita a limpeza dos dados para verificar na planilha, se existem dados sem informações ou com textos fora dos padrões. Classifica a coluna de números e informa ao phyton como modificar os números para que ele possa ler e trabalhar no programa corretamente.

```python
# LIMPEZA E TRATAMENTO - PADRONIZANDO OS DADOS PARA ANÁLISE

print("\n" + "=" * 50)
print("SPRINT 3 - LIMPEZA DE NULOS E DUPLICATAS")
print("=" * 50)

# Verificando as colunas vazias
colunas_vazias = df_tratado.columns[df_tratado.isnull().all()]
print("\nColunas completamente vazias encontradas:")
for coluna in colunas_vazias:
    print(f"- {coluna}")
df_tratado = df_tratado.drop(columns=colunas_vazias)
print("\nColunas vazias removidas com sucesso.")

#Verificando os valores nulos 
print("\nQuantidade de valores nulos por coluna:")
nulos = df_tratado.isnull().sum()
print(nulos)

#Tratando os valores nulos
total_nulos = nulos.sum()

if total_nulos == 0:
    print("\nNenhum valor nulo encontrado nas colunas úteis da base.")
else:
    print(f"\nForam encontrados {total_nulos} valores nulos.")


#Verificando os registros duplicados
duplicados = df_tratado.duplicated().sum()   #NECESSÁRIO REMOVER OS DUPLICADOS, PODEM SER DUAS COMPRAS.. VERIFICAR
print(f"\nQuantidade de registros duplicados: {duplicados}")


#Remoção dos duplicados
if duplicados > 0:
    
    linhas_antes = len(df_tratado)

    df_tratado = df_tratado.drop_duplicates()

    linhas_depois = len(df_tratado)

    removidos = linhas_antes - linhas_depois

    print(f"Foram removidos {removidos} registros duplicados.")

else:
    
    print("Nenhum registro duplicado foi encontrado.")                                

```

```python
print("Linhas originais:", len(df))
print("Linhas após limpeza:", len(df_tratado))

print(df_tratado[df_tratado.duplicated()].shape)
```

Foram identificados 96.553 registros completamente duplicados. Como os registros apresentavam igualdade em todas as colunas da base, optou-se pela remoção dessas duplicidades para evitar distorções nas análises estatísticas e nos agrupamentos posteriores. foi feita a conversão de data, datas invalidas 0, colunas removidas 4, valores nulos 0, duplicados removidos 96.553.
