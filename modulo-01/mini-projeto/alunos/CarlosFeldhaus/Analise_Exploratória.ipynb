Importando biblioteca

```python
import pandas as pd
```

Lendo arquivo

```python
df = pd.read_csv("Base Varejo.csv",sep=";", encoding= "latin1")
```

Buscando informações

```python
print(df.head())
```

Identificado colunas em branco, limpando.

```python
df.columns
df = df.dropna(axis=1, how='all')
df.info()
```

Nova visualização

```python
print(df.head())
```

Covertendo a Data

```python
df["DATA"] = pd.to_datetime(df["DATA"],format="%d/%m/%Y")
print(df["DATA"].dtype)
```

Verificando Nulos

```python
nulos = df.isnull().sum()

print("VALORES NULOS POR COLUNA")
print(nulos)
```

Duplicados

```python
print(f"Duplicados: {df.duplicated().sum()}")
```

```python
df = df.drop_duplicates()
```

```python
df["DATA"].isnull().sum()
```

Começando análises

```python
print(df["CL_FHL"].describe())
```

```python
compras_genero = (df.groupby("CL_GENERO").size().sort_values(ascending=False))

print(compras_genero)
```

```python
compras_categoria = (df.groupby("PR_CAT").size().sort_values(ascending=False))

print(compras_categoria)
```

```python
compras_segmento = (df.groupby("CL_SEG").size().sort_values(ascending=False))

print(compras_segmento)
```
