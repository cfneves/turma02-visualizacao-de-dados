**3 - LIMPEZA DOS DADOS**

```python
# Listando informações gerais sobre a base de dados antes do tratamento da data:
print(df.info())
```

```python
# Convertendo data para o formato datetime:
df["DATA"] = pd.to_datetime(df["DATA"], format="%d/%m/%Y")
```

```python
# Listando informações gerais sobre a base de dados depois do tratamento da data:
print(df.info())
```

```python
# Removendo as colunas Unnamed 10, 11, 12 e 13 devidos as mesmas não terem nenhum dado:
df.drop(columns=["Unnamed: 10", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"], inplace=True)
```

```python
# Listando informações gerais sobre a base de dados depois de remover as colunas Unnamed 10, 11, 12 e 13:
print(df.info())
```

```python
# Exibindo duplicatas e mostrando as 5 primeiras colunas duplicadas:
print(f"Duplicatas: {df.duplicated().sum()}")
display(df[df.duplicated()].head(5))
```

```python
# Removendo as linhas duplicadas:
df = df.drop_duplicates()
```

```python
# Exibindo duplicatas e mostrando as 5 primeiras colunas duplicadas:
print(f"Duplicatas: {df.duplicated().sum()}")
display(df[df.duplicated()].head(5))
```

**GERAR ESTATÍSTICAS DESCRITIVAS**

```python
# criando estatísticas descritivas para a coluna de filhos por cliente (CL_FHL) com graficos no matplotlib:

plt.figure(figsize=(10, 6))
plt.hist(df["CL_FHL"], bins=range(0, df["CL_FHL"].max() + 2), edgecolor="black")
plt.title("Distribuição do Número de Filhos por Cliente")
plt.xlabel("Número de Filhos")
plt.ylabel("Frequência")
plt.show()

# 1. Gerando a maior parte das estatísticas descritivas básicas
estatisticas = df['CL_FHL'].describe()
estatisticas['mediana'] = df['CL_FHL'].median()
estatisticas['moda'] = df['CL_FHL'].mode()[0]

# 3. Exibindo os resultados de forma organizada
print("=== Estatísticas Descritivas Básicas (Número de Filhos) ===")
print(f"Contagem (Total de registros): {estatisticas['count']:,.0f}")
print(f"Média:                         {estatisticas['mean']:.2f}")
print(f"Mediana:                       {estatisticas['mediana']:.0f}")
print(f"Moda:                          {estatisticas['moda']:.0f}")
print(f"Desvio Padrão:                 {estatisticas['std']:.2f}")
print(f"Mínimo:                        {estatisticas['min']:.0f}")
print(f"Máximo:                        {estatisticas['max']:.0f}")


```
