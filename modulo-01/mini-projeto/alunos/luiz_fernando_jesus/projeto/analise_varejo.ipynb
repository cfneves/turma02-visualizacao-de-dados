# Inicio do Sprint 3

```python
# Criando uma cópia do DataFrame Original
df_limpar = df.copy()

# Removendo as colunas Unnamed: 10, 11, 12 e 13.
df_limpar = df_limpar.drop(
    columns=["Unnamed: 10", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"]
)

# exibindo o DataFrame sem as colunas Unnamed: 10, 11, 12 e 13.
print("=== DATAFRAME SEM AS COLUNAS UNNAMED: 10, 11, 12 E 13 ===")
print(df_limpar.head().to_string())
```

```python
# importando uma nova biblioteca para tratar as valores nulos.
import numpy as np

# Transformando a string texto '#N/D' em nulo real (NaN).
df_limpar["PR_CAT"] = df_limpar["PR_CAT"].replace("#N/D", np.nan)
df_limpar["PR_NOME"] = df_limpar["PR_NOME"].replace("#N/D", np.nan)

# Preenchendo os novos NaNs com o termo correto (NÃO INFORMADO)
df_limpar["PR_CAT"] = df_limpar["PR_CAT"].fillna("NÃO INFORMADO")
df_limpar["PR_NOME"] = df_limpar["PR_NOME"].fillna("NÃO INFORMADO")

# Conferindoo resultado
print("=== VALIDAÇÂO DO PREENCHIMENTO ===")
print(f"{df_limpar['PR_CAT'].value_counts().to_string()}")
print(f"\n{df_limpar['PR_NOME'].value_counts().to_string()}")
```

```python
# Resolvendo o Problema da coluna DATA:
print(f"Tipo da Data antes: {df_limpar['DATA'].dtype}")
print(df_limpar["DATA"].head().to_string())
print()

# to_datetime para converter.
df_limpar["DATA"] = pd.to_datetime(
    df_limpar["DATA"], format="%d/%m/%Y", errors="coerce"
)

print(f"Tipo da Data depois: {df_limpar['DATA'].dtype}")
# Exibindo as primeiras linhas da coluna DATA:
print(df_limpar["DATA"].head().to_string())
```

```python
# Total de duplicatas antes da limpeza
print(f"Antes : {len(df_limpar)} linhas")

# Removendo as duplicatas já identificadas anteriormente.
df_limpar = df_limpar.drop_duplicates().reset_index(drop=True)

# Total depois da limpeza:
print(f"Depois: {len(df_limpar)} linhas")
print(f"Removidas: {len(df) - len(df_limpar)} duplicatas")
print(f"Duplicatas restantes: {df_limpar.duplicated().sum()}")
```

```python
# Resolvendo o ultimo problema detectado: Letras minusculas perdidadas no Texto.
# Formatando todas as colunas de str para letras maiusculas:

colunas_textuais = ["CL_GENERO", "CL_SEG", "PR_CAT", "PR_NOME"]

for col in colunas_textuais:
    df_limpar[col] = df_limpar[col].str.upper()

print("=== TEXTOS CONVERTIDO EM MAIUSCULO ===")

print(df_limpar.head(5).to_string())
```

```python
# Salvando a base o df_limpar em um arquivo csv.

df_limpar.to_csv(
    'Base_Varejo_Limpa.csv',
    index=False,
    sep=';',
    encoding='cp1252'
)

print('Base_Varejo_Limpa salvo com sucesso/!')
print(f'Colunas totais: {df_limpar.columns.tolist()}')
print(f'Linhas totais: {len(df_limpar)}')
```

# Inicio Sprint 4

```python
print('=== ANALISE DE NÚMERO DE FILHO DOS CLIENTE ===')

# COLUNA CL_FHL
coluna_filhos = df_limpar['CL_FHL']

# Coletando os parêmetros;
media = coluna_filhos.mean()
mediana = coluna_filhos.median()
desvio_padrao = coluna_filhos.std()
moda = coluna_filhos.mode()[0] # pega o primeiro valor da moda
maximo = coluna_filhos.max()
minino = coluna_filhos.min()
contagem = coluna_filhos.count()

# exibindo os relatórios
print(f'1. Contagem = Total de registros: {contagem}')
print(f'2. Mínimo de filhos: {minino}')
print(f'3. Máximo de filhos: {maximo}')
print(f'4. Média de filhos: {media:.2f}')
print(f'5. Mediana de filhos: {mediana}')
print(f'6. Moda de filhos: {moda}')
print(f'7. Devios padrao: {desvio_padrao:.2f}')
```

```python
# Analise por agupamento:
print('=== ANALISE DE AGRUPAMENTO 1: COMPRAS POR GÊNERO ===')

print(
    df_limpar.groupby('CL_GENERO')['CO_ID']
    .count()
    .sort_values(ascending=False)
    .to_string()
)

print('\n=== VOLUME DE COMPRAS POR CATEGORIA DE PRODUTO ===')

print(
    df_limpar.groupby('PR_CAT')['PR_ID']
    .count()
    .sort_values(ascending=False)
    .to_string()
)
```
